# Minesweeper

```python
#Generalised code for taking input for the grid size and placing multiple bombs #"\U0001F4A3"
```

```python
import random
def GenerateMineSweeperMap(n,m):
    
    arr1=[[0 for column in range(n)] for row in range(n)]

    for k in range(m):
        x= random.randint(0,n-1)
        y = random.randint(0,n-1)


        arr1[x][y]="X"
        
        if(y>=1): 
            if(arr1[x][y-1]!="X"):
            
                arr1[x][y-1]+=1   # middle left
    
        if(y<=n-2):             # middle right
            if(arr1[x][y+1]!="X"):
                
                arr1[x][y+1]+=1
    
        if(x>=1):             # top middle
            if(arr1[x-1][y]!="X"):
                
                arr1[x-1][y]+=1
    
        if(x<=n-2):             #bottom middle
            if(arr1[x+1][y]!="X"):
                arr1[x+1][y]+=1
    
        if(x>=1 and y>=1):   #top left
            if(arr1[x-1][y-1]!="X"):
                arr1[x-1][y-1]+=1
    
        if(x>=1 and y<=n-2):          #top right
            if(arr1[x-1][y+1]!="X"):
                arr1[x-1][y+1]+=1   
    
        if(y>=1 and x<=n-2):           #bottom left
            if(arr1[x+1][y-1]!="X"):
                arr1[x+1][y-1]+=1
    
        if(x<=n-2 and y<=n-2):           #bottom right
            if(arr1[x+1][y+1]!="X"):
                arr1[x+1][y+1]+=1
    
       
    return arr1 
        
    
    
    
    
```

```python
def GeneratePlayerMap(n):
    arr=[["-" for row in range(n)] for column in range(n)]
    return arr
```

```python
def DisplayMap(map):
    
    for row in map:
        
        print(" ".join(str(cell) for cell in row))
        print("")
```

```python
def CheckPlaying(map):
    for row in map:
        for c in row:
            if c=="-":
                return True
    return False        
```

```python
def CheckContinueGame(score):
    print("your score is: ", score)
    print("Do you Want to play again? Enter Y for yes and N for no")
    ch=input()
    if(ch=="Y" or ch=="y"):
        return True
    elif(ch=="N" or ch=="n"):
        return False
```

```python
def Game():
    GameStatus=True
    while GameStatus:
        print('''Difficulty levels:
                1. easy, grid size is 5 and number of bombs 3
                2. intermediate, grid size is 6 and number of bombs is 9
                3. difficult, grid size is 8 and number of bombs is 25
                ''')
        print("enter your choice as 1 or 2 or 3 ")
        ch=int(input())
        if(ch==1):
            n=5
            k=3
            
        elif(ch==2):
            n=6
            k=9
        elif(ch==3):
            n=8
            k=25
            
        minesweeper_map=GenerateMineSweeperMap(n,k)
        player_map=GeneratePlayerMap(n)
    
        score=0
    
        while True:
            if CheckPlaying(player_map):
                ch=int(input('''1. open a cell?
                                2. Flag a bomb?'''))
                if ch==1:
                    print("Enter the cell you want to open :")
                    x=input("x(1 to n(size of grid)): ")
                    y=input("y(1 to n(size of grid)): ")
                    x=int(x)-1
                    y=int(y)-1
                    if(minesweeper_map[x][y]=="X"):
                        print("GAME OVER!!")
                        DisplayMap(minesweeper_map)
                        GameStatus=CheckContinueGame(score)
                        break
                    else:
                        player_map[x][y]=minesweeper_map[x][y]
                        DisplayMap(player_map)
                        score+=1
                else:
                    print("enter the cell that you want to flag as bomb: ")
                    x=input("x(1 to (size of grid): )")
                    y=input("y(1 to (size of grid): )")
                    x=int(x)-1
                    y=int(y)-1
                    player_map[x][y]="X"
                    DisplayMap(player_map)
            elif(player_map==minesweeper_map):
                DisplayMap(player_map)
                print("wowwwwww!!!! You won !!")
                Gamestatus=CheckContinueGame(score)
                break
        
            else:
                print("GAME OVER!!!!")
                DisplayMap(minesweeper_map)
                GameStatus=CheckContinueGame(score)
                break
    print("your score is: ", score)       
            
            
            
        
```
